```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
### read dataset
```

```python
data = pd.read_csv('heart.csv')
```

```python
data.head(10)
```

```python
data.shape
```

```python
data.describe()
```

```python
y = data['target']
X = data[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
x_1 = sm.add_constant(X_train)
reg_log = sm.Logit(y_train, x_1)
results_log = reg_log.fit()
results_log.summary()
```

```python
x_2 = X_train.drop(['age','trestbps','chol','fbs','restecg','thalach','slope'], axis=1)

x_2 = sm.add_constant(x_2)
reg_log2 = sm.Logit(y_train, x_2)
result_log2 = reg_log2.fit()
result_log2.summary()
```

```python
result_log2.pred_table()
```

```python
### accuracy
```

```python
cm_df = pd.DataFrame(result_log2.pred_table())
cm_df.columns = ['Predicted 0', 'Predicted 1']
cm_df = cm_df.rename(index={0:'Actual 0', 1:'Actual 1'})
cm_df
```

```python
### accuracy train
```

```python
cm = np.array(cm_df)
accuracy_train = (cm[0,0] + cm[1,1]) / cm.sum()
accuracy_train
```
